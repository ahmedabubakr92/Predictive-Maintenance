## Maintenance & Its Types

__Maintenance__ is defined by the norm EN 13306 as the combination of ll technical, administrative and managerial actions during the life cycle of an item intended to retain it in., or restore it to, a state in which it can perform the required function. There are three form of machine maintenance to keep equipment running well:
1. _Corrective Maintenance (Run to Failure)_: is when maintenace is performed only after a machine fails. 
2. _Preventive Maintenance (Time Based or Scheduled)_: is when maintenance is performed on a regular basis according to a schedule. 
3. _Predictive Maintenanance (Proactive)_: is a maintenance strategy that improves efficiency by allowing a prediction of the remainng reusable life of the machinery also known as proactive maintenance. 

With developments in sensing and internet  of things technology, the importance in predictive maintenance has grown in recent decades. The key benefit in predictive maintenance is that it reduces unexpected downtime and associated expenses while also extending teh eqruipments life span. Predictive maintenance may be made easire using a variety of ways, using historical data and analyzing it with machine learning algorithms is one possible way. Because of the increased capability of processing power for sensors and failure data acquired in recent years, machine learning approaches are now quicker than traditional statistical survival analysis.

## Condition Monitoring: Inspection vs. Sensor-Based

How can data be gathered within condition monitoring? First and foremost, inspection based monitoring is the strategy that recieves the least attention when it comes to monitoring  a machine's status and collecting important data. With inspection based monitoring, data is gathered at regular intervals. The intervals are changed in accordance with observed and collected data on the present and projected conditional state of a machine or component. When a machine or a component is in its present and expected conditional state, the intervals are adjusted in accoradance with the observations and data acquired about the machine or component. Second, since advances in sensor technolog have made sensors for many sorts of metrics more affordable, the bulk of studies have focused on sensor-based monitoring various kindsof sensors such as those that detect vibration, temperature and acosutic emissions are utilized to gather the necessary information in sensor-based monitoring systems. It is important to note that sensor-based monitoring and inspection-based monitoring are not mutually exclusive since sensor equipment is often required in order to complete the inspection. In general, the use of sensor technology is more suited for an integrated predictive maintenance since it is critical for effective continuous monitoring and is more affordable. A machine or a component's remaining usable life may be estimated with the use of continuous monitoring, which collected important monitoring data throughout time. When compared to inspection-based monitoring, the quantity of data obtained is much more  since inspection-based monitoring is only a periodic snapshot of the conditional state of a machine. 

## Condition Monitoring in Practice

Although condition monitoring can be uased to predict failures, condition monitoring alone cannot be considered predictive because it only allows for the identification of parameter changes that occur before failure, it doesn't allow for the prediction of a relatively narrow future period of time during which parameter changes occur and thus the failure may occur. 

The inability to schedule maintenance activities in a timely manner prevents people responsible from doing efficient maintenance tasks such as determining the most cost-effective time frame for shutting down the manufacturing process for maintenance activities. So, a good prognostic method is needed to turn the datathat was collected into useful information for predicting when things will go wrong.

On the other hand, corrective maintenance is always done after there is a problem, following that the machine can either be repaired immediately or at a later time. This strategy, on the other hand, may result in a significant reduction in manufacturing time as well as costly maintenance. 

Reactive maintenance and corrective maintenance are two terms for the same thing. Preventive maintenance, also known as proactive maintenance, aims to keep things running smoothly. As a result, preventive maintenance policies are classified into two groups. First, preventive maintenance, which involves performing tasks at pre-determined intervals based on historicl average equipment life. On the other hand, condition-based maintenance continuously examines a machine's present state and plans maintenance tasks accordingly. It is possible to use three different types of condition-based monitoring in this situation, request-based monitoring, scheduled monitoring and continuous monitoring. Condition-based monitoring, on the otehr hand, keeps an eye on a machine's current state and plans maintenance actions around it. Monitoring on demand, planned monitoring and continuous monitoring are all viable options here. The first two methods are primarily inspection-based whereas continuous monitoring is usually accomplished through the use of sensors. 

## Benefits of Predictive Maintenance

The first benefit of predictive maintenance is that it could decrease the number of spare parts that need to be kept in stock. A reliable predictive maintenance system would make it possible to only keep the spare parts that will be needed soon instead of keeping every spare part that could be needed. With less spare parts in store and a smaller overall storage space, the predictive maintenance system can keep the maintenance process running smoothly and prevent production downtime caused by a lack of readily available replacement parts.

Secondly, some believe taht predictive maintenance might actually help extending the total life of machines and its components. Because predictive maintenance monitors a machine's health and anticipates its remaining usable life, it decreases the danger of a fatal breakdown, which might shorten the machine's lifespand even with thorough corrective maintenance. As a result, because predictive maintenance systems can help to avoid needless maintenance activities, they are also beneficial in that they can help to extend the life of machines. This is because every maintenance activity puts the machine at risk because a maintenance action and replaced components represent an intervention in the machine's structural design. 

Third, predictive maintenance systems are capable of detecting faults in the future based on monitoring dta and prognostic methodologies, the frequency of fatal breakdowns can be cut in half. There is always the possibility that a machine or component will fail prior to the next scheduled maintenancinterval even when using a conventional scheduled maintenances policy based on the historical data about the lifespand of a machine or a component. After all, the maintenance interval is only an average that includes outliers. A predictive maintenance strategy on the otehr hand, monitors the machines and its components on a frequent or even continuous basis and prompts maintenance activities regardless of whether or not the machine or a component is in use. As an alternative, a predictive maintenance strategy periodically or even continuously checks for malfunctions in the machine itself and its components and takes action, regardless of how long the equipment or component has been in use. 

## Cost Sensitive Machine Learning

Cost sensitive learning when it comes to predictive maintenance systems, training is the process of identifying the model parameter the best describe or explain the observed device behaviour. A cost function which compares actual and expected conditions for all devices to measure model performance is commonly used to determine best fit. When training a machine learning model, cost sensitive learning takes into consideration the costs of prediction mistakes as well as possibly other expenses. It is common to think of the predictive maintenance task as a binary classification problem because a device can be either broken or fully operation in a certain amount of time. 

The accuracy, recall and F1 score are standard cost functions for binary classification that are based on the confusion matrix. However, the most significant drawback of these confusion matrix-based performance indicators is their lack of understanding of the cost and impact of the predictive maintenance strategy on the organization. 

Important factors for a viable predictive maintenance strategy depend on the business case and include possible expenses for ticket diagnosis and triage , field trip and repair, components and shipment, as well as unavailability. 

We treat predictive maintenance asa  binary classification issue since a machine may be classified as functioning or imoperative during a certain forecast period. When we think of predictive maintenace as a binary classification issue its because a machine meay be classified as either functional or imoperative depending on how often it is predicted to work.

In many predictive maintenance applications, there's a mistmatch between maximizing F1 score and minimizing maintenance cost. It is not always desirable to misclassify an example from the majority class as an example from the minority class known as a _False Positive_ but it is less critical than misclassifying an example from the minority class as belonging to the majority class known as a _False Negative_ which is more serious.

## Data Description

Since real predictive maintenance datasets are generally difficult to obtain and in particular difficult to publish, we present and provide a synthetic dataset that reflects real predictive maintenance encountered in the industry to the best of our knowledge.

The dataset consists of 10 000 data points stored as rows with 14 features in columns

- UID: unique identifier ranging from 1 to 10000
- productID: consisting of a letter L, M, or H for low (50% of all products), medium (30%), and high (20%) as product quality variants and a variant-specific serial number
- air temperature [K]: generated using a random walk process later normalized to a standard deviation of 2 K around 300 K
- process temperature [K]: generated using a random walk process normalized to a standard deviation of 1 K, added to the air temperature plus 10 K.
- rotational speed [rpm]: calculated from powepower of 2860 W, overlaid with a normally distributed noise
- torque [Nm]: torque values are normally distributed around 40 Nm with an Ïƒ = 10 Nm and no negative values.
- tool wear [min]: The quality variants H/M/L add 5/3/2 minutes of tool wear to the used tool in the process. and a 'machine failure' label that indicates, whether the machine has failed in this particular data point for any of the following failure modes are true.

The machine failure consists of five independent failure modes
tool wear failure (TWF): the tool will be replaced of fail at a randomly selected tool wear time between 200 â€“ 240 mins (120 times in our dataset). At this point in time, the tool is replaced 69 times, and fails 51 times (randomly assigned).
heat dissipation failure (HDF): heat dissipation causes a process failure, if the difference between air- and process temperature is below 8.6 K and the tool's rotational speed is below 1380 rpm. This is the case for 115 data points.
power failure (PWF): the product of torque and rotational speed (in rad/s) equals the power required for the process. If this power is below 3500 W or above 9000 W, the process fails, which is the case 95 times in our dataset.
overstrain failure (OSF): if the product of tool wear and torque exceeds 11,000 minNm for the L product variant (12,000 M, 13,000 H), the process fails due to overstrain. This is true for 98 datapoints.
random failures (RNF): each process has a chance of 0,1 % to fail regardless of its process parameters. This is the case for only 5 datapoints, less than could be expected for 10,000 datapoints in our dataset.

If at least one of the above failure modes is true, the process fails and the 'machine failure' label is set to 1. It is therefore not transparent to the machine learning method, which of the failure modes has caused the process to fail

The data that is described above can be found linked here in [UCI Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/AI4I+2020+Predictive+Maintenance+Dataset).

### Feature Engineering

Since machine learning won't differentiate between the type of failures and will only conclude that there is a failure when one of the these features (TWF, HDF, PWF, OSF or RNF) indicate a failue. I'd like to create a feature that indicate the type of failue whenevr there's a machine failure. It will also help prevent overfitting will help create a better machine learning model.

```python
# Creating failure types columns with a str format
df['Tool Wear Failure'] = np.where(df['TWF'] == 1, 'Tool Wear Failure', df['TWF'])
df['Heat Dissipation Failure'] = np.where(df['HDF'] == 1, "Heat Dissipation Failure", df['HDF'])
df['Power Failure'] = np.where(df['PWF'] == 1, 'Power Failure', df['PWF'])
df['Overstrain Failure'] = np.where(df['OSF'] == 1, 'Overstrain Failure', df['OSF'])
df['Random Failure'] = np.where(df['RNF'] == 1, 'Random Failure', df['RNF'])
```

```python
# Concatenating the columns created to create a failure type column
df['Failure_type'] = df['Tool Wear Failure'] + df['Heat Dissipation Failure'] + df['Power Failure'] + \
df['Overstrain Failure'] + df['Random Failure']

# Checking the value counts in the "Failure_type" column
df['Failure_type'].value_counts()
```

```python
# Lets clean and fix the failure type column
df['Failure_type'] = df['Failure_type'].replace({'00000':"No Failure", 
                                '0Heat Dissipation Failure000': 'Heat Dissipation Failure',
                               '00Power Failure00':'Power Failure',
                               '000Overstrain Failure0':'Overstrain Failure',
                               'Tool Wear Failure0000':'Tool Wear Failure',
                               '0000Random Failure':'Random Failure',
                               '00Power FailureOverstrain Failure0':'Power Failure/Overstrain Failure',
                               '0Heat Dissipation Failure0Overstrain Failure0':'Heat Dissipation Failure/Overstrain Failure',
                               '0Heat Dissipation FailurePower Failure00':'Heat Dissipation Failure/Power Failure',
                               'Tool Wear Failure00Overstrain Failure0':'Tool Wear Failure/Overstrain Failure',
                               'Tool Wear Failure000Random Failure':'Tool Wear Failure',
                               'Tool Wear Failure0Power FailureOverstrain Failure0':'Tool Wear Failure/Power Failure/Overstrain Failure'},
                                                   )
```

```python
# Checking the value counts in the "Failure_type" column
df['Failure_type'].value_counts()
```

### Cross Validation

For splitting the data into features and taregt, I'm going to use Stratified K-Fold which is a cross-validator that provides train/test indices to split data in train/test sets. This cross-validation object is a variation of KFold that returns stratified (formed or arranged into layers) folds. The folds are made by preserving the percentage of samples for each class.

It is applied as follows:

```python
# Variable with column names
col_names = df2.columns.to_list()

# Selecting features and targets
predictive_columns = col_names[:5]
X = df3[predictive_columns]
y = df3['Failure_type']
```

```python
# importing StratifiedKFold from sklearn
from sklearn.model_selection import StratifiedKFold

# Creating an instance of StartifiedKFold
skf = StratifiedKFold(n_splits=5, random_state=None)

# Applying the StratifiedKFold
for train_index, test_index in skf.split(X, y):
    X1_train, X1_test = X.iloc[train_index], X.iloc[test_index]
    y1_train, y1_test = y.iloc[train_index], y.iloc[test_index]
    train_0, train_1 = len(y1_train[y1_train == 0]), len(y1_train[y1_train == 0])
    test_0, test_1 = len(y1_test[y1_test == 0]), len(y1_test[y1_test == 1])
    print(f"Train: 0 = {train_0}, 1 = {train_1}, Test: 0 = {test_0}, 1 = {test_1}")
```

### Multi-Class & Multi-Label Classification

For further model building, we need to get an understanding of multi-label classification and how multi-class classification can be defined. 

For __Multi-label Classification__, the data must include more than one independent variable, also known as target class, and the cardinality of each class must be binary. There's more than one output prediction in this sort of categorization issue. The majority of classification machine learning algorithms are incapable of dealing with multi-label classification problems. To train multi-label classification data, a wrapper around machine learning techniques is required. This technqiue involves fitting one classifier to each target. This enables the categorization of numerous target variables. The goal of this class is to expand estimators such that they may estimate a set of target functions that have been trained on a single X predictor matrix to predict a set of answers. 

__Multi-class Classification__ can be categorized as a single output learning model when the output class is represented by the integer encoding. It can also be extended to a multi-output learning scenario if each output class is represented by the One Hot vector. The purpose of multi-output classification is to train a classification model which output is a collection of labels or vector. 

In comparison, the multi-label classification entails categorizing occurrences into numerous labels that share semantics such as the difficulty of caterozing movies according to genre, a movie may be classifier as eitehr action or thriller, but it can also be classified as both, for example, action/thriller, for instance, thriller and action movies share semantics, they both relate to the movie semantics and are thus two different values for the same movie. It is also likely that a movie cannot be classified as any previously known genre or that it is best classified as several movie genres. 

Building a predictive that simultaneously outputs a set of two or more labels that measure different concepts is thus the machine learning task of solving a multi-output problem, essentially, two or more separate but related classification problems are solved concurrently within the same model.

##  Multi-Class False Positives, Negatives and True Positives, Negatives

At this stage, I'd like to evaluate the models in terms of costs related to incorrect classified validation data of the created models. 

A Multi-Class classification model will consequently also have a confusion matrix extending the already known two by two format. Mostcategorization metrics are binary by default. The classification problems goal is to determine the the class to which an observtion belongs based on the collection of related characteristics. Typically, a collection of training datasets containing observations with known class mempership is given and classification is a subset of supervised learning in this context. A classification issue is classified as binary if the class label can only take two values, for example a or b, Multi-class if the class label can take several values, for example a, b, or c and Multi-label if each observation is connected with many classes. 

In my predictive maintenance case, we'll have a six by six confusion matrix for five failure types and class no failure. All the diagonal columns corresponding to each failure type are considered as true positive. The total number of true negatives for a failure type is the sum of the values of all the columns and rows excluding that failure type classes rows and columns. The total number of false positives for a class is the sum of the values of the corresponding rows excluding true positives. The total number of false negatives for a failure type is the sum of the values of the corresponding column excluding true positives.

## Summary of Multi-Label & Multi-Class Classifications

The previous trained Multi-Label classifiers with the function `MultiOutputClassifer` all have high accuracy, precision, recall and F1 score for failure types. Overall, Random Forest Classifeir and CatBoost perform slightly more robust and accurate compared to Light Gradient Boosted Machine and XGBoost for Multi-Class and Multi-Label Classification tasks. Anyhow, as with many other predictive maintenance tasks, cost sensitive learning is the main driver for classification model selection. 

The final assessment is therefore based on costs related to quantity of false positive and false negative classifications. As I have five different failure types, I need to consider that the overall maintenance costs differ among failure types. The cost impact related to false positive and false negative per failure type differs as well as false positives and false negatives related to class "No Failure". There'll be a need to consider that false negatives related to "No Failure" class as additionally to maintenance costs, an undetected failure can result in severe machine failure  and will cause costly production downtime.

Maintenance costs for false negatives of all failure type classes is the sum of repair costs plus replacement of part costs due to misclassification of failure type. The overall costs of true failure type will be considered in the cost calculation. For example, in case of failure type which is a power failure that was predicted as tool wear failure, maintenance costs will be calculated for power failure. 

Maintenance costs for false positives of all failure types takes replacement of part costs due to misclassification of failure type into account. The repair costs of the true failure type will be considered in the cost calculations as costs for part replacements have been already calculated by false negatives. This approach is because misclassification lead to wrongly assumed failure types and repair costs due to labor working hours are relevant. 

## Cost Sensitive Learning - CatBoost

After making cost calculations for the most robust and accurate classifiers, Random Forest, Light Gradient Boosted Machine and CatBoost, I'll take the metrics from CatBoost as example to understand the cost calculations.

Multi-Class Classifications as previously shown with the method MultiOutput Classifier have a Multi-Class confusion matrix which was described previously. The advantage of a Multi-Class confusion matrix comparing to a binary-based multi-label confusion matrix is that misclassifications to the remaining classes are shown in the matrix, within a binary confusion matrix misclassifications to other classes are not shown. For cost calculations, the Multi-Class with Output Classifier is preferred as cost allocations for false negatives and false positives. 

It can be seen here that the Multi-Class classification with One Vs. One CatBoost model with total cost of false negatives and false positives is less costly, keeping in mind that only 20 false negatives with no failure have been classified. That would mean that in 20 cases there'll be a machine failure and consequently production downtime without detecting a proper failure type.

Unplanned production downtime due to undetected failures is critical for business and also reliability of machine learning based predictive maintenance. In my calculations, the CatBoost algorithm was one of the cost optimal multi-class classification model and also Random Forest classifier was very cost optimal.

Severity of production downtime due to non-operational machines because of undetected failure types is multiple times higher and much more critical. In conclusion, due to the low false negative values for class "No Failure" and transparency of Multi-Class classification confusion matrix, it is recommended to select CatBoost Multi-Class classification model as model of choice for the predictive maintenance task.
